```python
##2.HIGHEST SAALRY FROM EMPLOYEE DATA
#taking input in dictionary format
employees={"Ravi":75000,"Anitha": 68000,"Kiran":72000}
#variable for highest salary and highest name
highest_salary=0
highest_name=""
#loop through salary
for name,salary in employees.items():
    if salary > highest_salary:
        highest_salary=salary
        highest_name=name
##printing result
print("Highest Salary:",highest_name,"-",highest_salary)


```

```python
##1.UNIQUE WORDS IN A SENTENCE
#Taking the sentence/input
sentence=input("Enter a sentence:")
#splitting the sentence into individual words
words=sentence.split()
#coverting list to set to remove duplicates
unique_words=set(words)
#for  printing count of unique_words
count=len(unique_words)
#printing the result
print("Unique words count:",count)
print("Unique words:",unique_words)
```

```python
##4.COUNT PRODUCTS ABOVE A PRICE THRESHOLD
prices=[450,1200,899,1500,300]
highest_price=1000
#intiallizing count
count=0
#checking every price like comparing with highest price
for price in prices:
  if price > highest_price:
    count=count+1
print("Products above 1000:",count)
```

```python
#3.FIND MAXIMUM AND MINIMUM VALUES
#list of numbers taking as input
numbers=[45,22,89,10,66]
#variables for storing max and min
max=numbers[0]
min=numbers[0]
#loop through list and checking every num in list
for num in numbers:
  if num > max:
    max = num
  if num < min:
    min = num
print("List:",numbers)
print("Max",max)
print("Min:",min)
```

```python
#7.COUNT CHARACTER FREQUENCY
#given text as input
text="python"
#take empty dict for frequency
frequency={}
#count every or each character usig for loop and if else
for char in text:
  if char in frequency:
    frequency[char] += 1
  else:
    frequency[char] = 1
print(frequency)
```
