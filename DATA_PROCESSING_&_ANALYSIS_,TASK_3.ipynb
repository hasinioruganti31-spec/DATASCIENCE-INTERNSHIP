```python
#1..Employee Performance Bonus Eligibility
#given input
employees = {
    "Ravi:": 92,
    "Anitha": 88,
    "Kiran": 92,
    "Suresh":85
}
#highest_score intializing
highest_score = max(employees.values())
top_performers = []
#checking every employee
for name, score in employees.items():
  if score ==  highest_score:
    top_performers.append(name)
#final result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers),f"(Score: {highest_score})")

```

```python
#5.SALES SPIKE DETECTION
sales = [1200, 1500, 900, 2200, 1400, 3000]
total=0
A = len(sales)
for amount in sales:
  total = total + amount
#calculating avg
avg = total/A
#checking every day
day = 1
for i in sales:
  if i > avg:
    print("Day", day, ":", i)
  day = day + 1
```

```python
#6.DUPLICATE USER ID DETECTION
#input given
user_ids = ['user1',"user2","user1","user3","user1","user3"]
id_count = {}
#count
for uid in user_ids:
  if uid in id_count:
    id_count[uid] += 1
  else:
    id_count[uid] = 1
#printing by checking
for uid, count in id_count.items():
  if count > 1:
    print(f"{uid} -> {count} times")
```

```python
#4.EMAIL DOMAIN USAGE ANALYSIS
#input given
emails = ["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]
domain_count = {}
#checking or extracting through loop
for email in emails:
  domain = email.split("@")[1]
  domain_count[domain] = domain_count.get(domain, 0) + 1
total = len(emails)
#calculating percentage of domain
for domain, count in domain_count.items():
  percentage = (count / total) * 100
  print(f"{domain}:{int(percentage)}%")
```

```python
#2.SEARCH QUERY KEYWORD ANALYSIS
#input given
query = "Buy mobile phone buy phone online"
#coverting the input to lowercase
query = query.lower()
#slpitting the words
words = query.split()
#count to store
word_count = {}
for word in words:
  if word in word_count:
    word_count[word] += 1
  else:
    word_count[word] = 1
#storing repeted words
result ={}
for word, count in word_count.items():
  if count > 1:
    result[word] = count
print(result)
```
