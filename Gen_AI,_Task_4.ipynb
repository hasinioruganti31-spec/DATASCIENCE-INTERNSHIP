```python
#1.MOBILE RECHARGE VALIDATION SYSTEM
def validate_recharge():
  #list of plans
  valid_plans=[199,299,399,599]
  #checking through while loop
  while True:
    #taking recharge amount as input
    amount=int(input("Enter recharge amount: "))
    if amount >=50 and amount in valid_plans:
      print("Recharge Succesful")
      break
      #exit loop if recharge is valid->break
    else:
      print("Invalid recharge.Try again.")
      #function calling
validate_recharge()
```

```python
#2.INVENTORY REORDER ALERT SYSTEM
def check_inventory(stock_data):
  #checking through loop
  for product, quantity in stock_data.items():
    if quantity < 15:
      print(product, ": Reorder Alert")
    else:
      print(product, ": Stock OK")
      #dict of products storing in it
products = {"Rice":12,"Oil":20,"Sugar":8,"Wheat":30}
#calling function
check_inventory(products)
```

```python
#3.STUDENT RESULT PROCESSING SYSTEM
def student_result(marks):
  #intilaizing total marks
  total=0
  L=len(marks)
  #checking through loop
  for mark in marks:
    total = total+mark
  average = total/L
  if average >=50:
    return "Pass"
  else:
    return "Fail"
    #list of students
marks_list=[11,12,11,1]
#calling a function
result=student_result(marks_list)
#result
print("Result:",result)
```

```python
#5.EMPLOYEE ATTENDENCE ATTENDENCE ELIGIBILITY CHECKER
def attendence_check(attendence):
  present_days=0
  L=len(attendence)
  #checking through loop ,counting number of presentees"P"
  for day in attendence:
    if day =="P":
      present_days += 1
  percentage = (present_days/L)*100
  if percentage >= 75:
    return "Eligible"
  else:
    return "Not Eligible"
    #list of attendence
attendence_list=["P","A","P","P","A","P","P"]
#calling a function
status = attendence_check(attendence_list)
print("Attendence Status:",status)
```

```python
#4.CAB FARE ESTIMATOR WITH RETRY OPTION
def calculate_fare(distance, peak):
  base_fare = 50
  cost = base_fare + (distance * 12)
  if peak == "yes":
    cost += cost *0.25
  return cost
while True:
  #input
  km = float(input("Enter distance in km: "))
  peak_time = input("Is it a peak time? (yes/no): ").lower()
  #calculating fare
  fare = calculate_fare(km, peak_time)
  print("Total Fare: ", int(fare))
  retry = input("Do you want to calculate again? (yes/no): ")
  if retry != "yes":
    #breaks the loop when we say no
    break
```

```python
#6PASSWORD STRENGTH CHECKER
def password_strength(password):
  #checking digit
  has_digit = False
  #checking special character
  has_special = False
  special_characters ="@#$&"
  #checking through loop
  if len(password) < 8:
    return "Weak Password"
  for ch in password:
    if ch.isdigit():
      has_digit = True
    if ch in specual_characters:
      has_special = True
  if has_digit and has_special:
    return "Strong Password"
  else:
    return "Weak Password"
    #input
password=input("Enter password: ")
#result
print(password_strength(password))
```
