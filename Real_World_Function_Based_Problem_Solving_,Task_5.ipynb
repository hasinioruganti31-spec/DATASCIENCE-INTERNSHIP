```python
#Problem 1: Social Media – Post Engagement Analyzer
#defining a fuction
def analyze_post_engagement(likes_list):
  total_likes=0
  #loop to calculate the total likes
  for likes in likes_list:
    total_likes+=likes
    #condition th check post status
  if total_likes>=1000:
    status="Viral Post"
  else:
    status="Normal Engagement"
  print("Total likes:",total_likes)
  print("Post Status:",status)
likes_list=[200,350,400,100]
#calling a function
analyze_post_engagement(likes_list)

```

```python
#Problem 3: Agriculture – Rainfall Adequacy Checker
def Rainfall_adequacy_checker(rainfall_list,required_level):
  total=0
  #calculating total rainfall
  for rainfall in rainfall_list:
    total=total+rainfall
  average=total/len(rainfall_list)
  if rainfall >= required_level:
    Rainfall_Status="Adequate Rainfall"
  else:
    Rainfall_Status="Inadequate Rainfall"
  print("Average Rainfall:",average)
  print("Rainfall Status:",Rainfall_Status)
Rainfall_list=[70,75,68,80,67]
required_level=70
  #calling a function
Rainfall_adequacy_checker(Rainfall_list,required_level)
```

```python
#Problem 6: Agriculture – Premium Crop Price Filter
def premium_crops(prices):
  premium_list=[]
  #loop through prices in the list
  for price in prices:
    if price>2000:
      premium_list.append(price)
  print("Premium Crops:",premium_list)
  #example data according to the output given in the question
prices=[1500,2500,3200,900]
premium_crops(prices)

```

```python
#4.Social Media – Duplicate Account Detection
#defining a function
def detect_duplicate_accounts(usernames):
  #converting list of usernmes to sets to remove duplicates
  unique_names=set(usernames)
  if len(usernames) != len(unique_names):
    result="Yes"
  else:
    result="No"
  print("Duplicate Accounts Found:",result)
usernames=["hasini","hasini_sithara","Hasini","hasini"]
detect_duplicate_accounts(usernames)

```

```python
#Problem 7: System Monitoring – Application Health Checker
# defining a function
def health_checker(errors):
  #checking through conditional statements
  if errors==0:
    status="Healthy"
  elif errors <= 5:
    status="Minor Issues"
  else:
    status="Critical Issues"
  print("Error Count:",errors)
  print("System Status:",status)
health_checker(7)
```

```python
#8.Banking – Daily Transaction Limit Checker
#defining a function
def Transaction_limit_checker(amount):
  #given limit
  daily_limit=50000
  #checking through conditionals
  if amount <= daily_limit:
    status="Approved"
  else:
    status="Rejected"
  print("Transaction Amount:",amount)
  print("Transaction Status:",status)
  #calling a function
Transaction_limit_checker(60000)
```

```python
#10.Smart Electricity Bill Analyzer
#defining function
def analyse_bill(units):
  #bill calculation
  if units <= 100:
    bill = units * 3
  elif units <= 200:
    bill=(100*3)+(units-100)*5
  else:
    bill=(100*3)+(100*5)+(units-200)*7
      #usage classification
  if bill < 500:
    usage = "  Low Usage"
  elif bill <= 1500:
    usage ="Moderate Usage"
  else:
    usage="High Usage"
  return bill, usage
#calling afunction
amount, usage=analyse_bill(250)
print("Total Bill:",amount)
print("usage Status:",usage)
```
