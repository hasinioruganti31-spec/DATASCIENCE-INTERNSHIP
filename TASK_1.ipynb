```python
##3.Simple Data Cleaner:
names=["Alice","bob","CHARLIE"]
cleaned_names=[]
#checking
for name in names:
  #REMOVE
    n=name.strip()
  #CONVERTING TO LOWERCASE
    n=n.lower()
  #adding names into list
    cleaned_names.append(n)
print("Cleaned names List:", cleaned_names)

```
