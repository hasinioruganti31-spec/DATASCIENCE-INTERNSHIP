```python
#Problem 1: Smart Light Controller
#defining class
class Smart_light_controller:
  def __init__(self,light_name):
    self.light_name=light_name
    self.status="OFF"
  def turn_on(self):
    self.status="ON"
    #displaying the status
  def show_status(self):
    print(f"{self.light_name} is {self.status}")
    # taking input here
light_name=input("Light Name:")
action=input("Action:")
#creating object
light=Smart_light_controller(light_name)
if action.upper()=="ON":
  light.turn_on()
light.show_status()
```

```python
#3.Mobile Contact Record
class Contact:
  def __init__(self,name,phone):
    self.name=name
    self.phone=phone
  def show_contact_details(self):
    print("Contact Saved")
    #here iam using f string to print you can use in other way also
    print(f"Name: {self.name}")
    print(f"Phone: {self.phone}")
#input
name=input("Contact Name:")
phone=input("Phone Number:")
#creating object here
contact=Contact(name, phone)
contact.show_contact_details()
```

```python
#4.Product Price Tag
class Product:
  def __init__(self,product_name,price):
    self.product_name=product_name
    self.price=price
  #generating here price tag
  def generate_price_tag(self):
    print(f"Product:{self.product_name}")
    print(f"Price: {self.price}")
    #input given
product_name=input("Product Name:")
price=input("Price:")
#creating object
product=Product(product_name,price)
product.generate_price_tag()
```

```python
#5Movie Rating Display System
#defining a class
class System:
  def __init__(self,movie_name,rating):
    self.movie_name=movie_name
    self.rating=rating
    #showing details here
  def show_movie_details(self):
    print(f"Movie: {self.movie_name}")
    print(f"Rating: {self.rating}/5")
    #input
movie_name=input("Movie Name: ")
rating = input("Rating:")
#creating object
movie=System(movie_name,rating)
#printing result
movie.show_movie_details()
```

```python
#6.Delivery Address Manager
#defining a class
class Delivery:
  def __init__(self,name,address):
    self.name=name
    self.address=address
  def print_delivery_details(self):
    print("Delivery Details")
    print(f"Customer: {self.name}")
    print(f"Address: {self.address}")
name=input("Customer Name:")
address=input("Address:")
#creating object
delivery=Delivery(name,address)
#printing result
delivery.print_delivery_details()
```
