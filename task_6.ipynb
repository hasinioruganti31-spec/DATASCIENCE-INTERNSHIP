```python
#1.Smart Parking Lot Management System
#defining a function
def parking_system(capacity,logs):
  current_count=0
  peak_count=0
  for action in logs:
    if action == "IN":
      current_count+=1
    elif action =="OUT":
      current_count-=1
  #tracking maximum vehicles parked
    if current_count>peak_count:
      peak_count=current_count
  print("Currently Parked Vehicles:",current_count)
 #checking parking status
  if current_count>=capacity:
    print("Parking Status: Full")
  else:
    print("Parking Status: Available")
parking_capacity=50
vehicle_logs=["IN","IN","IN","OUT","IN","IN","OUT"]
#calling a function
parking_system(parking_capacity,vehicle_logs)
```

```python
#2.Online Food Delivery Time Estimator
#defining function
def estimate_delivery(distance,traffic, weather):
  #as base time 5 min per km
  base_time=distance*5
  #traffic delay conditions
  if traffic == "High":
    base_time+=15
  elif traffic == "Medium":
    base_time+=10
  #weather delay condition
  if weather =="Stormy":
    base_time+=15
  print("Estimated Delivery Time:",base_time,"minutes")
  #calling a function
estimate_delivery(8,"High","Rainy")
```

```python
#3.Movie Theatre Seat Occupancy Analyzer
#defining a function
def seat_occupancy(total_seats,booked_list):
  booked_count=len(booked_list)
  occupancy=(booked_count/total_seats)*100
  print("Occupancy:",int(occupancy),"%")
  if occupancy==100:
    print("Show Status: Housefull")
  elif occupancy>=75:
    print("Show Status:Alomost Full")
  else:
    print("Show Status:Available")
#sample input we are taking here
total_seats=200
booked_seats=[1]*150
seat_occupancy(total_seats,booked_seats)
```

```python
#4.Cloud Server Load Classification System
#defining a function
def server_load(cpu_readings):
  total=0
  #loop
  for value in cpu_readings:
    total+=value
  average = total/len(cpu_readings)
  print("Average CPU Load:",int(average),"%")
  #classification based on rules
  if average < 50:
    print("Server Status: Normal")
  elif average<80:
    print("Server Status:Warning")
  else:
    print("Server Status:Critical")
  #give sample data
cpu_readings=[45,60,70,85,90]
server_load(cpu_readings)
```

```python
#5.Smart Classroom Resource Usage Monitor
def resource_monitor(usage_data):
  #list of overused resouces storing here
  overused=[]
  #checking
  for resource, hours in usage_data.items():
    if hours > 8:
      overused.append(resource)
      if overused:
        print("overused Resources:",", ".join(overused))
        print("Energy Alert: Yes")
      else:
        print("Overused Resources: No")
#sample input
resources={"Projector":6,"AC":9,"Lights": 4}
resource_monitor(resources)
```

```python
#6.Online Event Registration Capacity Controller
#defining a function
def event_registration(capacity,total_registrations):
  if total_registrations <= capacity:
    print("Confirmed Registraions:",total_registrations)
    print("Waitlisted Users: 0")
    print("Registration Status: Open")
  else:
    waitlist=total_registrations-capacity
    print("Confirmed Registrations:",capacity)
    print("Waitlisted Users:",waitlist)
    print("Registration Status:Closed")
#sample input
event_capacity=100
registrations=105
#calling
event_registration(event_capacity,registrations)
```
